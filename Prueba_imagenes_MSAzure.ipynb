#### Programa para probar Cognitive Services de MS Azure para deteccion de caras en imagenes


```python
import json, os, requests
```

```python
#Asignamos la clave de subscripcion del servicio de MS Azure
subscription_key='324d928aff81414e848f36101bf110ca'

#declaramos la ruta donde se encuentra nuestra API activa en la nube
face_api_url= 'https://facerecognition-a0035169.cognitiveservices.azure.com/'+'face/v1.0/detect/'

#le pasamos la url de la imagen que vamos a usar ejemplo1
image_url='https://raw.githubusercontent.com/ecendejas/imgMNA/main/faces.jpg?raw=true'

# Le pasamos el diccionario con la llave de susbcripcion al headers para cuando hagamos el request
headers={'Ocp-Apim-Subscription-Key': subscription_key}
```

```python
# Le pasamos los parametros que utilizara la API de deteccion
params= {'returnFaceID': 'false', 'returnFaceLandmarks': 'false'}
```

```python
# hacemos el request para la API con los parametros para detectar la imagen
response = requests.post(face_api_url,params=params, headers=headers, json={'url': image_url})
print (json.dumps(response.json()))
```

**Esto nos entrega los resultados de las coordenadas donde detecta cada cara en la imagen**
